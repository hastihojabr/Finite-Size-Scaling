## 1. Simulation using Metropolis Algorithm

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
from IPython.display import display, HTML
warnings.filterwarnings("ignore")
```

```python
def Ising(N,Ti,Tf,n,equlibration_step,measurment_step): 
    
    #Lattice Size , initial Temperature , final Temperature , Temperature steps , Equlibration Step , Measurment Step
    
    def Random_Matrix(N):  
        return np.random.choice([1,-1], size=(N,N))

    def Metropolis(State,t,N,step):
        m,mm,ee,e=[],[],[],[]
        Eng,EngEng,Mag,MagMag = [],[],[],[]
        for s in range(step):
            En=0
            for i in range(N*N):
                x, y = np.random.randint(0,N) , np.random.randint(0,N)
                Neighbour = State[(x-1)%N,y]+State[x,(y-1)%N]+State[(x+1)%N,y]+State[x,(y+1)%N]
                dE = 2*State[x,y]*Neighbour
                if dE<0:
                    State[x,y] =- State[x,y]       
                elif np.random.random(1)[0] < np.exp(-(beta[t])*dE):
                    State[x,y] =- State[x,y]
                En -= State[x,y]*Neighbour  
                
            M = np.sum(State) 
            m.append(abs(M))
            mm.append(abs(M)**2)
            e.append(En/2)
            ee.append(En**2/4)
            print(abs(M),En**2/4)
        
        Mag,MagMag,Eng,EngEng = np.mean(m),np.mean(mm),np.mean(e),np.mean(ee)
        
        C=(beta[t]**2)*(EngEng-Eng**2)/(N**2)
        x = (beta[t])*(MagMag-Mag**2)/(N**2)
        return State , Eng/(N**2) , C , sum(m)/step/(N**2) , x
    
    E,X,Cv,M,U = np.zeros(n),np.zeros(n),np.zeros(n),np.zeros(n),np.zeros(n)
    T = np.linspace(Ti,Tf,n) 
    beta = 1/T

    for t in range(0,n):
        State = Random_Matrix(N) 
        Metropolis(State,t,N,equlibration_step)                   #equlibration
        State,e,cv,m,x = Metropolis(State,t,N,measurment_step)    #measurment

        E[t],M[t],Cv[t],X[t] = e,m,cv,x

    fig = plt.figure(figsize=(18, 10))
    name = ['Energy','Magnetization','Specific Heat','Susceptibility']

    for i,ax,name in zip([E,abs(M),Cv,X],range(1,5),name):
        ax =  fig.add_subplot(2, 2, ax )
        ax.scatter(T,i,color = 'k',marker = '.')
        ax.set_xlabel('Temperature')
        ax.set_ylabel(name)

    return plt.show(),E,Cv,M,X,T
```

##### 2.1 Finding Critical Temperature Using the peak in (T-Cv) and (T-X) graphs

```python
def Poly_fit(x,y,degree):
    fig = plt.figure(figsize=(15, 5))
    op = []
    for d in range(1,degree):
        plt.subplot(1,2,1)
        plt.scatter(x,y,color='r',marker='.')
        fit = np.poly1d(np.polyfit(x ,y, d))
        plt.plot(x,fit(x),label='Polynomial Degree: {}'.format(d))
        plt.legend()
        coeffs = np.polyfit(x, y, d)
        p = np.poly1d(coeffs)
        yhat = p(x)
        ybar = np.sum(y)/len(y)
        ssreg = np.sum((yhat-ybar)**2)
        sstot = np.sum((y - ybar)**2)
        results = 1 - (((1-(ssreg/sstot))*(len(y)-1))/(len(y)-d-1))
        op.append(results)
    df = pd.DataFrame({'Polynomial Degree': [i for i in range(1,degree)], 'r_squared': op})
    Degree = np.argmax(df.r_squared)+1
    model = np.poly1d(np.polyfit(x,y,Degree))
    plt.subplot(1,2,2)
    plt.scatter(x,y,color='r',marker='.',label='Monte Carlo Data')
    Data = model(x)
    plt.plot(x, model(x), color='k',label='Fitted Data')
    #plt.title(name)
    i = np.argmax(Data)
    return plt.show,df, Data , i
```

```python
def Tc_1(T,Q,degree):
    _,df,data,i = Poly_fit(T,Q,degree+1)
    Tc = T[i]
    loss = (abs(Tc-2.26918531421)/2.26918531421)*100
    plt.subplot(1,2,2)
    plt.axvline(x=Tc, color='k', linestyle='-',label="Model's Tc")
    plt.axvline(x=2.26918531421, color='k', linestyle='-.',label="Actual Tc")
    plt.legend()
    print('Critical Tempreture is :{} with accuracy of :{}'. format(Tc,100-loss) )
    return 
```

#### 2.2  The Binder Cumulant

```python
def Binder(N,Ti,Tf,n,equlibration_step,measurment_step): 
    
    #Lattice Size , initial Temperature , final Temperature , Temperature steps , Equlibration Step , Measurment Step
    
    def Random_Matrix(N):  
        return np.random.choice([1,-1], size=(N,N))

    def Metropolis(State,t,N,step):
        mm,mmmm=[],[]
        MagMag,MagMagMagMag = [],[]
        
        for s in range(step):
            En=0
            for i in range(N*N):
                x, y = np.random.randint(0,N) , np.random.randint(0,N)
                Neighbour = State[(x-1)%N,y]+State[x,(y-1)%N]+State[(x+1)%N,y]+State[x,(y+1)%N]
                dE = 2*State[x,y]*Neighbour
                if dE<0:
                    State[x,y] =- State[x,y]       
                elif np.random.random(1)[0] < np.exp(-(beta[t])*dE):
                    State[x,y] =- State[x,y] 
            M = np.sum(State)
            
            mm.append(abs(M)**2)
            mmmm.append(abs(M)**4)
            
        MagMag, MagMagMagMag  = np.mean(mm), np.mean(mmmm)
        U=(1-(MagMagMagMag)/(3*MagMag**2))

        return State  , U
    
    U = np.zeros(n)
    T = np.linspace(Ti,Tf,n) 
    beta = 1/T 
    
    for t in range(0,n):
        State = Random_Matrix(N) 
        Metropolis(State,t,N,equlibration_step)            #equlibration
        State,u = Metropolis(State,t,N,measurment_step)    #measurment
        U[t] = u
    fig = plt.figure(figsize=(5, 5))
    plt.scatter(T,U,color = 'k',marker = '.')
    return plt.show(),U,T
```

#### 2.3 Real Space Renormalization Group

```python
def RG_Ising(N,Ti,Tf,n,equlibration_step,measurment_step,stride): 
    
    #Lattice Size , initial Temperature , final Temperature , Temperature steps , Equlibration Step , Measurment Step
    
    def Random_Matrix(N):  
        return np.random.choice([1,-1], size=(N,N))

    def majority_rule(Q,stride):
        def submatrix(matrix, startRow, startCol, size):
            return matrix[startRow:startRow+size , startCol:startCol+size]
        blocked = np.zeros((len(Q)//stride,len(Q)//stride))
        for i in range(0,len(Q),stride):
            for j in range(0,len(Q),stride):
                s = submatrix(Q, i, j, stride)
                blocked[i//stride,j//stride] = s[0,0]
        return blocked
    
    def u_RG(State,N,stride):
        state = majority_rule(State,stride)
        u,NN,ulist=0,N//stride,[]
        for i in range(NN**2):
            x, y = np.random.randint(0,NN) , np.random.randint(0,NN)
            spin = state[x,y]
            Neighbour= state[(x-1)%NN,y]+state[x,(y-1)%NN]+state[(x+1)%NN,y]+state[x,(y+1)%NN]
            u -= state[x,y]*Neighbour
        ulist.append(u/2)
        return NN , ulist
    
    def calcEnergy(config):
        'Energy of a given configuration'
        energy,E = 0,[]
        for i in range(len(config)):
            for j in range(len(config)):
                energy += -config[i,j]*(config[(i+1)%N, j] + config[i,(j+1)%N] + config[(i-1)%N, j] + config[i,(j-1)%N])
        E.append(energy/2)
        return E
    
    def Metropolis(State,t,N,step):
        Eng,e=[],[]
        for s in range(step):
            En=0
            for i in range(N*N):
                x, y = np.random.randint(0,N) , np.random.randint(0,N)
                Neighbour = State[(x-1)%N,y]+State[x,(y-1)%N]+State[(x+1)%N,y]+State[x,(y+1)%N]
                dE = 2*State[x,y]*Neighbour
                if dE<0:
                    State[x,y] =- State[x,y]       
                elif np.random.random(1)[0] < np.exp(-(beta[t])*dE):
                    State[x,y] =- State[x,y]
            e = calcEnergy(State)
            NN,ulist = u_RG(State,N,stride)
        Eng = sum(e)/step
        UEng = sum(ulist)/step
        return State , Eng/(N**2) , UEng/(NN**2)
    
    E,U = np.zeros(n),np.zeros(n)
    T = np.linspace(Ti,Tf,n) 
    beta = 1/T 
    
    for t in range(0,n):
        State = Random_Matrix(N) 
        Metropolis(State,t,N,equlibration_step)                   #equlibration
        State,e,u = Metropolis(State,t,N,measurment_step)        #measurment
        E[t],U[t]= e,u

    fig = plt.figure(figsize=(5, 5))
    name = ['Energy','Energy RG']
    plt.plot(T,E,color = 'k',label=name[0])
    plt.plot(T,U,color = 'r',label=name[1])
    plt.legend()
    return plt.show(),E,U
```
